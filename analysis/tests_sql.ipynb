```python
import pandas as pd
```

```python
path = '../data/'
transitions = pd.read_csv(path+'transitions.csv',sep=',')
```

```python
transitions = transitions[transitions['to_state']=='booked']
spaces = transitions[['space_id','city','category']]
spaces.columns = ['id','city','space_type']
bookings = transitions[['booking_request_id','city','space_id','idea_id',
                        'start_on','end_on','price']]
bookings.columns = ['id','city','space_id','idea_id',
                        'start_on','end_on','price_pounds']
bookings.to_csv('bookings_sql.csv')
spaces.drop_duplicates('id').to_csv('spaces_sql.csv')
```

```python
sql_string ="SELECT strftime('%Y', end_on) AS year, strftime('%m', end_on) AS month, city, SUM(price_pounds) AS revenue FROM (SELECT * FROM bookings WHERE price_pounds > 5000)  GROUP BY year, month, city"
df = pd.read_sql(sql_string, conn)
df
```

```python
sql_string ="SELECT , strftime('%Y', end_on) AS year, strftime('%m', end_on) AS month, city, SUM(price_pounds) AS revenue FROM bookings  GROUP BY year, month, city  HAVING revenue > 5000"
df = pd.read_sql(sql_string, conn)
df
```
